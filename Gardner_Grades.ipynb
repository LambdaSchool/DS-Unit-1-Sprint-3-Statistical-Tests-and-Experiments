```
d = []
subject = 'Afrikaans'
for index, row in grade_oneA.iterrows():
  if row["Afrikaans"] <= 70:
    d.append([row["Student Number"], row[subject], "Low Understanding"])
  elif row["Afrikaans"] <= 85:
    d.append([row["Student Number"], row[subject], "Mid Understanding"]) 
  elif row["Afrikaans"] <= 100:
    d.append([row["Student Number"], row[subject], "High Understanding"])

df = pd.DataFrame(d)


df.rename(columns={0:'Student Number', 1:'Average', 2:'Understanding'}, inplace=True)

pd.crosstab(df['Understanding'], df['Student Number'], margins=True)



```
