```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import bokeh
```

```python
pops = pd.read_csv('US_Populations_by_State_edited2 09-20-30-291.csv')
delta = pd.read_csv('us percent change.csv')
```

```python
pops['Confederacy'] = pops['TX'] + pops['AR'] + pops['LA'] + pops['TN'] + pops['MI'] + pops['AL'] + pops['GA'] 
+ pops['FL'] + pops['SC'] + pops['NC'] + pops['VA']

pops.head()
```

```python
pops['Union'] = pops['MA'] + pops['NY'] + pops['NH'] + pops['VT'] + pops['MA'] + pops['CT'] + pops['RI'] 
+ pops['PA'] + pops['NJ'] + pops['OH'] + pops['IN'] + pops['IL'] + pops['KS'] + pops['MI'] + pops['WI'] 
+ pops['MN'] + pops['IA'] + pops['CA'] + pops['NV'] + pops['OR']

pops['Border'] = pops['MD'] + pops['DE'] + pops['WV'] + pops['KY'] + pops['MO']
```

```python
delta['Confederacy'] = delta['TX%'] + delta['AR%'] + delta['LA%'] + delta['TN%'] + delta['MI%'] + delta['AL%']
+ delta['GA%'] + delta['FL%'] + delta['SC%'] + delta['NC%'] + delta['VA%']

delta['Union'] = delta['MA%'] + delta['NY%'] + delta['NH%'] + delta['VT%'] + delta['MA%'] + delta['CT%'] + delta['RI%'] + delta['PA%'] + delta['NJ%'] + delta['OH%'] + delta['IN%'] + delta['IL%'] + delta['KS%'] + delta['MI%'] + delta['WI%'] 
+ delta['MN%'] + delta['IA%'] + delta['CA%'] + delta['NV%'] + delta['OR%']

delta['Border'] = delta['MD%'] + delta['DE%'] + delta['WV%'] + delta['KY%'] + delta['MO%']

delta.head()
```

```python
cw_delta = delta.drop(columns=['AL%','AR%','AZ%','CA%','CO%','CT%','DC%','DE%','FL%','GA%','IA%','ID%',
                      'IL%','IN%','KS%','KY%','LA%','MA%','MD%','ME%','MI%','MN%','MO%','MS%',
                      'MT%','NC%','ND%','NE%','NH%','NJ%','NM%','NV%','NY%','OH%','OK%','OR%',
                      'PA%','RI%','SC%','SD%','TN%','TX%','UT%','VA%','VT%','WA%','WI%','WV%',
                      'WY%'])
```

```python
cw_delta.head()
```

```python
cw_pops = pops.drop(columns=['AL','AR','AZ','CA','CO','CT','DC','DE','FL','GA','IA','ID',
                      'IL','IN','KS','KY','LA','MA','MD','ME','MI','MN','MO','MS',
                      'MT','NC','ND','NE','NH','NJ','NM','NV','NY','OH','OK','OR',
                      'PA','RI','SC','SD','TN','TX','UT','VA','VT','WA','WI','WV',
                      'WY'])

cw_pops.head()
```

```python
cw_pops.describe()
```

```python
# data for plotting

year = cw_pops['Year']
pops_C = cw_pops['Confederacy']
pops_U = cw_pops['Union']
pops_B = cw_pops['Border']

fig, ax = plt.subplots()

# lines

ax.plot(year, pops_C)
ax.plot(year, pops_B)
ax.plot(year, pops_U)

ax.set(xlabel='year', ylabel='population (millions)',
       title="Confederacy, Union, & Border States' Population")

plt.ylim(5000000,70000000)
plt.yticks(np.arange(5000000,75000000,step=5000000),('5','10','15','20','25','30','35','40',
                                                    '45','50','55','60','65','70'))

plt.legend()

# ax.grid()

plt.show();

fig.savefig('USAPopulation.png')
```
