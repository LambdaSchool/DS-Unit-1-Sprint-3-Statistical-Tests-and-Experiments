```
class Gaussian:
  def __init__(self, m, s, p=1):
    self.mu = m+0.000001
    self.sgm = s+0.000001
    self.p  = p
  
  def skalar(self): 
    return self.p / (self.sgm * sqrt(2 * pi))
  
  def rad_denom(self):
    return 2.0 * (self.sgm ** 2) 
  
  def normal(self, x):
    k = self.skalar()
    r = self.rad_denom()
    the_pow = (0-1) * (((x - self.mu)**2) / r)
    nrm = k * (e ** the_pow)
    return nrm

  def chart(self): 
    l = 5
    x = np.linspace(-l,l,100)
    y = self.normal(x)
    display = pd.DataFrame({'x': x, 'y': y})
    
    C = alt.Chart(display).mark_line(
        # marker='crosshairs', 
        tooltip='mu={}\nsigma={}'.format(self.mu, self.sgm)
    ).encode(
      x='x',
      y='y'
    ).properties(
      height=100,
      width=100    
    )
    return C
```

```
from scipy.special import gamma, hyp2f1
from math import sqrt, pi, e
from functools import reduce
import numpy as np
import pandas as pd
import altair as alt
# https://en.wikipedia.org/wiki/Student%27s_t-distribution
# for cdf: https://en.wikipedia.org/wiki/Hypergeometric_function 
```

```
class StudentT:
  def __init__(self, degrees_of_freedom):
    self.v = degrees_of_freedom
  
  def pdf(self, x):
    skalar_numer = gamma((self.v + 1)/2.)
    skalar_denom = sqrt(self.v * pi) * gamma(self.v / 2)
    phi = (1 + (x**2/self.v))
    powwa = - (self.v + 1) / 2
    return (skalar_numer / skalar_denom) * (phi**powwa)
  
  def cdf(self, x):
    fact1 = x * gamma((self.v+1)/2)
    fact2 = hyp2f1(0.5, (self.v+1)/2, 1.5, x**2/self.v)
    fact3 = (sqrt(pi*self.v) * gamma(self.v/2))**(-1)
    return 0.5 * fact1 * fact2 * fact3
  
  def chart(self): 
    l = 5
    x = np.linspace(-l,l,100)
    y_pdf = self.pdf(x)
    y_cdf = self.cdf(x)
    display_pdf = pd.DataFrame({'x': x, 'y': y_pdf})
    y_cdf = self.cdf(x)
    display_cdf = pd.DataFrame({'x': x, 'y': y_cdf})
    
    pdf_line = alt.Chart(display_pdf).mark_line(
        # marker='crosshairs', 
        tooltip='v={}\nprobability density function'.format(self.v),
        color='red'
    ).encode(
      x='x',
      y='y'
    ).properties(
      height=100,
      width=100    
    )
    cdf_line = alt.Chart(display_cdf).mark_line(
        tooltip='v={}\cumulative density function'.format(self.v),
        color='green'
    ).encode(
      x='x',
      y='y'
    ).properties(
      height=100,
      width=100    
    )
    return pdf_line & cdf_line
```
