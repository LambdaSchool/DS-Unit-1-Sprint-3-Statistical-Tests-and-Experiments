```
import pandas as pd 
import scipy
from scipy import stats

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data', sep=',', header=None, skiprows=1)
data.tail(30)

```

```
data = data.replace('-', '0')
data = data.replace('+', '1')

data.tail(10)

```

```
ttest = scipy.stats.ttest_1samp(data[15], 0.5)
print(ttest)
```

```
approved = data.loc[data[15] == 1]
denied = data.loc[data[15] == 0]

stats.ttest_ind(approved[15], denied[15])
```

```
data_compare = pd.DataFrame({'approved': approved[15],
                   'denied': denied[15]})

data_compare.head()
```
